```python
#short question
"""
                    Vth
                |           |
---------- R1 ------- R2 ----
|                          |
V                         R3
|                          |
----------------------------- 
"""

import random

def thevenin_generator():
    V = random.randint(1, 23)
    R1 = random.randint(1000, 11000)
    R2 = random.randint(1000, 11000)
    R3 = random.randint(1000, 11000)

    #vth
    vth = V * R2/(R1 + R2 + R3)

    #rth
    rth = R1 + R3

    print(f"Voltage Source: {V} V")
    print(f"Resistor 1: {R1} ohms, Resistor 2: {R2} ohms, Resistor 3: {R3} ohms")
    print(f"Thevenin Voltage: {vth:.2f} V, Thevenin Resistance: {rth:.2f} ohms")

thevenin_generator()






```

```python
#long question

"""
    Vb
    ___________ R __________    Va 
    |                       |
    |           Vc          |
    |---R-------|-----R-----|-----------|  ^
 |  I1          R           R           I2 |
 v  |___________|___________|___________|

"""

def circuit():
    R1 = random.randint(1000, 11000)
    R2 = random.randint(1000, 11000)
    R3 = random.randint(1000, 11000)
    R4 = random.randint(1000, 11000)
    R5 = random.randint(1000, 11000)
    I1 = random.randint(1, 10)
    I2 = random.randint(1, 10)

    A = np.array([
        [1/R2 + 1/R4 + 1/R5, -1/R2, -1/R4],
        [1/R4, 1/R1, -1/R1 - 1/R3 - 1/R4],
        [1/R2, -1/R1 - 1/R2, 1/R1]
    ])

    # Define the b vector
    b = np.array([I2, 0, I1])

    va, vb, vc = np.linalg.solve(A, b)

    return va, vb, vc

va, vb, vc = circuit()
print(f"Voltage across I1: {vb}")
print(f"Voltage across I2: {va}")
```

```python
L = 101*10**(-3) #the problem with this is that the signal doesnt attenuate very much
C = 100*10**(-6)
R = 10000
w = smp.symbols('w')
s = smp.symbols('s') 
"""where s is jw"""
r, l, c = smp.symbols('r l c')

Xl = s*L
Xc = 1/(s*C)

Xt = (Xl * Xc)/(Xl + Xc)

Hw = R / (R + Xt)

Hw
```

```python
Hw_simplified = smp.simplify(Hw)

num, den = smp.fraction(Hw_simplified)
num_coeffs = smp.Poly(num, s).all_coeffs()
den_coeffs = smp.Poly(den, s).all_coeffs()

num_coeffs = [float(coeff) for coeff in num_coeffs]
den_coeffs = [float(coeff) for coeff in den_coeffs]

cct = signal.TransferFunction(num_coeffs, den_coeffs)

w, mag, phase = signal.bode(cct)

#Plot magnitude
plt.figure()
plt.figure(figsize=(12, 6))
plt.semilogx(w/(2*smp.pi), mag)
plt.title('Bode Plot - Magnitude')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude [dB]')
plt.grid()

# Plot phase
plt.figure()
plt.figure(figsize=(12, 6))
plt.semilogx(w/(2*smp.pi), phase)
plt.title('Bode Plot - Phase')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Phase [degrees]')
plt.grid()

plt.show()
```

```python
def f(x, num_terms=50):
    result = 2.5
    for n in range(1, num_terms + 1):
        term = (5 * (1 - (-1)**n)) / (n * np.pi) * np.sin(2000 * n * np.pi * x)
        result += term
    return result

x = np.linspace(0, 10, 1000)

y = f(x, num_terms=50)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Series Approximation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the Series Function')
plt.legend()
plt.grid(True)
plt.show()
```

The history of power plug design:

For New Zealand the power plug is based on the historic wye plug, which was patented in the U.S 1915. The same design is used in Australia, Argentina, and China. The U.S has since changed to a different design due to a rule change where lighting circuits were allowed to be fused up to 15A for circuits under 125V but stayed at 10A over 125V. This peculiar rule lead to a wall plug change and a deviation from the rest of the world.

Some plugs have 2 prongs because plugs only require the phase and neutral connection. Plugs that have three prongs have the third prong there as an earth connection for safety. if the neutral connection breaks, the connection to ground acts as a safety return for the circuit.
